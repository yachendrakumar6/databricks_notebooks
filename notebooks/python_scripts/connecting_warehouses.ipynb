```python
    # Set up your AWS credentials
    aws_access_key = 'YOUR_AWS_ACCESS_KEY'
    aws_secret_key = 'YOUR_AWS_SECRET_ACCESS_KEY'
    mfa_serial_number = 'MFA_SERIAL_NUMBER'

    # Prompt to request a MFA token
    mfa_token = input("Enter your MFA token: ")

    # Set up Spark session
    spark = SparkSession.builder \
     .appName("S3_Parquet") \
     .config("spark.hadoop.fs.s3a.impl",  "org.apache.hadoop.fs.s3a.S3AFileSystem") \
     .config("spark.hadoop.fs.s3a.access.key", aws_access_key) \
     .config("spark.hadoop.fs.s3a.secret.key", aws_secret_key) \
     .config("spark.jars.packages", "org.apache.hadoop:hadoop-aws:3.3.2,com.amazonaws:aws-java-sdk-bundle:1.12.115") \
     .config("spark.hadoop.fs.s3a.aws.credentials.provider.params", f"mfa_serial={mfa_serial_number},mfa_token={mfa_token}") \
     .getOrCreate()


     # Read Parquet file from S3 using Spark
     file_path = "s3a://path/to/part- 00000-000-1111-4444-5555-6677889900.c111.snappy.parquet"

     df = spark.read.parquet(file_path)

     # Show data
     display(df)
```
