```python
from pyspark.sql.functions import *
from pyspark.sql.functions import *

silver_df = (
    df.select(
        coalesce(initcap(trim(col("emp_name"))), lit("")).alias("emp_name"),

        coalesce(upper(trim(col("department"))), lit("")).alias("department"),

        when(
            col("email").rlike("^[A-Za-z0-9._%+-]+@[A-Za-z0-9.-]+\\.(com|org|edu|gov|net|co|io|in)$"),
            col("email")
        ).otherwise(lit("")).alias("email"),

        coalesce(col("salary"), lit(0)).alias("salary"),
        col("status"),
        col("address")
    )
)

display(silver_df)
```
